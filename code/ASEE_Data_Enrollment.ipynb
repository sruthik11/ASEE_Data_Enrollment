```python
import pandas as pd

def filter_data(dataframe):
    # Filtering conditions
    major_mapping = {
        "ASEN": "1. Aerospace Engineering Sciences (B.S.)",
        "CBEN": "2. Chemical and Biological Engineering (B.S.)",
        "CHEN": "3. Chemical Engineering (B.S.)",
        "CSAP": "4. Applied Computer Science Post Bachelors (B.S.)",
        "CSEN": "5. Computer Science (B.S.)",
        "MCEN": "6. Mechanical Engineering (B.S.)",
        "TMEN": "7. Creative Technology and Design (B.S.)",
        "AMEN": "8. Applied Mathematics (B.S.)",
        "BMEN": "9. Biomedical Engineering (B.S.)",
        "EPEN": "10. Engineering Physics (B.S.)",
        "IDEN/GEEN": "11. Engineering Plus (B.S.)",
        "EVEN": "12. Environmental Engineering (B.S.)",
        "AREN": "13. Architectural Engineering (B.S.)",
        "CVEN": "14. Civil Engineering (B.S.)",
        "ECEN": "15. Electrical and Computer Engineering (B.S.)",
        "EEEN": "16. Electrical Engineering (B.S.)",
        "XXEN": "17. Open Option"
    }

    ethnicity_mapping = {
        "AMERIND": "1. American Indian",
        "ASIAN": "2. Asian",
        "BLACK": "3. Black",
        "HISPA": "4. Hispanic",
        "PACIF": "5. Pacific Islander",
        "WHITE": "6. White",
        "TWOPLUS": "7. Two or More",
        "NONRESALIEN": "8. Nonresident Alien",
        "UNKNOWN": "9. Unknown"
    }

    dataframe.loc[:, 'Major'] = dataframe['Major'].replace({'IDEN': 'IDEN/GEEN', 'GEEN': 'IDEN/GEEN'})
    # Create a DataFrame with all possible combinations of majors and ethnicity groups
    majors = list(major_mapping.keys())
    ethnicities = list(ethnicity_mapping.keys())
    all_combinations = pd.DataFrame([(major, ethnicity) for major in majors for ethnicity in ethnicities], columns=['Major', 'IPEDS_ETHNIC_GRP_CD'])

    # Merge the all_combinations DataFrame with the filtered DataFrame to ensure all majors are included
    merged_df = all_combinations.merge(dataframe, on=['Major', 'IPEDS_ETHNIC_GRP_CD'], how='left')
    merged_df.fillna(0, inplace=True)

    # Apply the conditions and group by major, ethnicity, gender, and student status
    filtered_df = merged_df.groupby(['Major', 'IPEDS_ETHNIC_GRP_CD', 'Gender_CD', 'SCD_FLPT']).size().reset_index(name='count')

    # Mapping
    filtered_df['Major'] = filtered_df['Major'].map(major_mapping)
    filtered_df['IPEDS_ETHNIC_GRP_CD'] = filtered_df['IPEDS_ETHNIC_GRP_CD'].map(ethnicity_mapping)

    # Pivot the DataFrame to create the desired format with new column names
    formatted_df = filtered_df.pivot(index=['Major', 'IPEDS_ETHNIC_GRP_CD'], columns=['SCD_FLPT', 'Gender_CD'], values='count')

    # Define the desired column order
    column_order = [
        ('F', 'M'), ('F', 'F'), ('P', 'M'), ('P', 'F')
    ]

    # Reorder the columns
    formatted_df = formatted_df[column_order]

    # Rename the columns to match the desired format
    formatted_df.columns = ['Full-Time Male', 'Full-Time Female', 'Part-Time Male', 'Part-Time Female']

    # Fill NaN values with 0
    formatted_df = formatted_df.fillna(0).astype(int)

    # Reset the index to keep "Major" and "IPEDS_ETHNIC_GRP_CD" as columns
    formatted_df.reset_index(inplace=True)

    formatted_df['Full-Time Total'] = formatted_df['Full-Time Male'] + formatted_df['Full-Time Female']
    formatted_df['Part-Time Total'] = formatted_df['Part-Time Male'] + formatted_df['Part-Time Female']
    formatted_df['Grand Total'] = formatted_df['Full-Time Total'] + formatted_df['Part-Time Total']

    # Reorder the columns to match the desired format
    formatted_df = formatted_df[['Major', 'IPEDS_ETHNIC_GRP_CD', 'Full-Time Male', 'Full-Time Female', 'Full-Time Total', 'Part-Time Male', 'Part-Time Female', 'Part-Time Total', 'Grand Total']]

    

    return formatted_df

def add_major_total_rows(dataframe):
    # Calculate the sum for each major
    major_total_rows = dataframe.groupby('Major').sum(numeric_only=True).reset_index()

    # Set the 'IPEDS_ETHNIC_GRP_CD' column in the total rows to 'Total'
    major_total_rows['IPEDS_ETHNIC_GRP_CD'] = 'Total'
    
    # Create a DataFrame to store the final result
    final_dataframe = pd.DataFrame(columns=dataframe.columns)
    
    # Iterate through majors and insert the total row for each major group
    for major in dataframe['Major'].unique():
        major_data = dataframe[dataframe['Major'] == major]
        major_total_data = major_total_rows[major_total_rows['Major'] == major]
        
        final_dataframe = pd.concat([final_dataframe, major_data, major_total_data], ignore_index=True)
        
        # Add a blank row with NaN values after the total row
        final_dataframe = final_dataframe.append(pd.Series([None] * len(final_dataframe.columns), index=final_dataframe.columns), ignore_index=True)
    
    return final_dataframe




```

```python
# Define a function to write each major's data to the file
def write_major_data(file, major, data):
    header = 'Major,Ethnicity,Full-Time Male, Full-Time Female, Full-Time Total, Part-Time Male, Part-Time Female, Part-Time Total, Grand Total\n'
    file.write(header)

    # Get unique major names
    major_names = data['Major'].unique()

    for major_name in major_names:
        major_data = data[data['Major'] == major_name]

        # Write the major name for each group of rows with the same major
        major_name_row = f'{major_name},,\n'
        file.write(major_name_row)

        # Remove the major name from the data and write the data
        data_without_major = major_data.copy()
        data_without_major['Major'] = ''
        data_without_major.to_csv(file, index=False, header=False, lineterminator='\n')

# Define a list of majors in the desired order
def custom_open(dataframe, csv_file_name):
    desired_major_order = [
    "1. Aerospace Engineering Sciences (B.S.)",
    "2. Chemical and Biological Engineering (B.S.)",
    "3. Chemical Engineering (B.S.)",
    "4. Applied Computer Science Post Bachelors (B.S.)",
    "5. Computer Science (B.S.)",
    "6. Mechanical Engineering (B.S.)",
    "7. Creative Technology and Design (B.S.)",
    "8. Applied Mathematics (B.S.)",
    "9. Biomedical Engineering (B.S.)",
    "10. Engineering Physics (B.S.)",
    "11. Engineering Plus (B.S.)",
    "12. Environmental Engineering (B.S.)",
    "13. Architectural Engineering (B.S.)",
    "14. Civil Engineering (B.S.)",
    "15. Electrical and Computer Engineering (B.S.)",
    "16. Electrical Engineering (B.S.)",
    "17. Open Option"
    ]

    # Save the DataFrame to a CSV file with custom formatting while maintaining the order
    with open(csv_file_name, 'w') as file:
        for major in desired_major_order:
            major_data = dataframe[dataframe['Major'] == major]
            if not major_data.empty:
                write_major_data(file, major, major_data)
                file.write('\n')  # Add an empty line to separate major sections
```

```python
import pandas as pd

# List of CSV files to convert to sheets
csv_files = ["freshman_data.csv", "sophomore_data.csv", "junior_data.csv", "senior_data.csv", "grad_data.csv"]

# Output Excel file
output_excel_file = "output.xlsx"

# Create a Pandas Excel writer
excel_writer = pd.ExcelWriter(output_excel_file, engine="xlsxwriter")

# Loop through the CSV files and convert them to Excel sheets
for csv_file in csv_files:
    sheet_name = csv_file.split(".")[0]  # Use the CSV file name as the sheet name
    df = pd.read_csv(csv_file)
    df.to_excel(excel_writer, sheet_name=sheet_name, index=False)

# Save the Excel file
excel_writer.save()
```
